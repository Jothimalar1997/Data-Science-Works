```python
'''Q5. Load the wine quality data set and perform exploratory data analysis (EDA) to identify the distribution of each feature. Which feature(s) exhibit non-normality, and 
what transformations could be applied tothese features to improve normality?'''
'''The following features exhibit non-normality:
(i) Residual sugar concentrations have a positive skewing that moves the mean value above the median. An extreme outlier can be found around 65 g/L residual sugar2.
(ii) Chlorides have a skewed distribution with a long tail on the right side3.
To improve normality, you could apply the following transformations:
(i) Logarithmic transformation for residual sugar concentrations2.
(ii) Square root transformation for chlorides'''
categorical_features=[feature for feature in df.columns if df[feature].dtype=='O']
numerical_features=[feature for feature in df.columns if df[feature].dtype!='O']
print("we have {} numerical features:{}".format(len(numerical_features),numerical_features))
print("we have {} categorical features:{}".format(len(categorical_features),categorical_features))
plt.figure(figsize=(15,15))
plt.suptitle('Univariate Analysis of Numerical Features',fontsize=20,fontweight='bold',alpha=0.8,y=1.)
for i in range(0,len(numerical_features)):
  plt.subplot(5,3,i+1)
  sns.kdeplot(x=df[numerical_features[i]],shade=True,color='r')
  plt.xlabel(numerical_features[i])
  plt.tight_layout()
```
