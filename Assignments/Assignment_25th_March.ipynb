```python
## Q13.How does the type of app affect its price? Create a bar chart to compare average prices by app type.
plt.figure(figsize=(15,6))
df['Price'] = df['Price'].apply(lambda x: float(x.replace('$', '')) if x != '0' and x != 'Everyone' else 0)
df.groupby('Type')['Price'].mean().plot(kind='bar')
plt.title('Average Prices by App Type')
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.show()
```

## Insights:
- If app type is '0' or 'Free', it gets more installations than Priced one.

```python
## Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.
category = pd.DataFrame(df['Category'].value_counts()) 
category.rename(columns = {'Category':'Count'},inplace=True)
plt.figure(figsize=(15,6))
print("Bar Plot")
sns.barplot(x=category.index[:10], y ='Count',data = category[:10],palette='hls')
plt.title('Top 10 App categories')
plt.xticks(rotation=90)
plt.show()
print("Frequency Table")
plt.figure(figsize=(15,16))
sns.heatmap(category,annot=True)
```
