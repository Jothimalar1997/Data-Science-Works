```python
## Q2. Define the different types of statistics and give an example of when each type might be used.
'''There are two types of statistics- Descriptive Statistics, Inferential Statistics
In the case of descriptive statistics, the data or collection of data is described in summary. 
But in the case of inferential stats, it is used to explain the descriptive one. Both these types have been used on large scale.
(i) Descriptive Statistics:
The data is summarised and explained in descriptive statistics. The summarization is done from a population sample utilising several factors such as mean and standard deviation. 
It is a way of organising, representing, and explaining a set of data using charts, graphs, and summary measures. Histograms, pie charts, bars, and scatter plots are common ways to summarise data and present it in tables or graphs. 
They don’t need to be normalised beyond the data they collect.

(ii) Inferential Statistics:
We attempt to interpret the meaning of descriptive statistics using inferential statistics. 
We utilise inferential statistics to convey the meaning of the collected data after it has been collected, evaluated, and summarised. 
The probability principle is used in inferential statistics to determine if patterns found in a study sample may be extrapolated to the wider population from which the sample was drawn. 
Inferential statistics are used to test hypotheses and study correlations between variables, and they can also be used to predict population sizes. 
It is used to derive conclusions and inferences from samples, i.e. to create accurate generalisations.

Example
In a class, the Data is the set of marks obtained by 50 students. Now when we take out the Data average, the result is the average of 50 students’ marks. It is the example of descriptive statistics.
If the average marks obtained by 50 students are 88 out of 100, on the basis of the outcome, we will draw a conclusion. It is the example of inferential statistics.'''
```

```python
## Q5. Explain the concept of levels of measurement and give an example of a variable for each level.
'''1) Nominal scale data:
It is measured on qualitative/ categorical variables.
Here, order does not matter. Eg: Gender, colors

2) Ordinal scale data:
Rank and order matters in these types of measurements.
The difference between them cannot be measured.
Eg: Qualification - PhD, BE, Masters, B.com

3) Interval scale data:
Rank and order matters in these types of measurements.
The difference between them can be measured excluding ratio.
It doesn't have 0 as it's starting value.
Eg: Temperature

4) Ratio scale data:
Rank and order matters in these types of measurements.
The difference and ratio can be measured.
It has 0 as it's starting value.
Eg: Grades'''
```

```python
## Q6. Why is it important to understand the level of measurement when analyzing data? Provide an example to illustrate your answer.
'''Levels of measurement are important because they indicate the extent to which statisticians, marketing analysts and financial analysts can use existing data.
For instance, if you classify your data at the ordinal level of measurement, you know that you can evaluate how items compare to each other based on a specific hierarchy.
If you wanted to analyze the exact distance between items rather than understand only how they compare to one another, you might want to rely on data at the interval level of measurement.
Understanding levels of measurement can also help you properly plan your research.
For instance, if you want to gather data with a true zero point, you can invest the appropriate resources to go beyond nominal data and gather data that adheres to the ratio level of measurement. 
The levels of data measurement are also important for providing insight into how to present research findings. 
Eg: when applying interval measurement, you might present your data in charts or graphs, whereas ratio measurement might use mathematical equations.'''
```

```python
## Q10. What are some common measures of central tendency and variability used in statistics? Explain how each measure can be used to describe a dataset.
'''Measures of central tendency help you find the middle, or the average, of a dataset. The 3 most common measures of central tendency are the mode, median, and mean.
(i) Mode: the most frequent value. The mode can be used for any level of measurement, but it’s most meaningful for nominal and ordinal levels.
(ii) Median: the middle number in an ordered dataset. The median can only be used on data that can be ordered – that is, from ordinal, interval and ratio levels of measurement.
(iii) Mean: the sum of all values divided by the total number of values. The mean can only be used on interval and ratio levels of measurement because it requires equal spacing between adjacent values or scores in the scale.

Variability describes how far apart data points lie from each other and from the center of a distribution. Along with measures of central tendency, measures of variability give you descriptive statistics that summarize your data.
Variability is also referred to as spread, scatter or dispersion. It is most commonly measured with the following:

(i) Range: the difference between the highest and lowest values.
For data measured at an ordinal level, the range and interquartile range are the only appropriate measures of variability.
(ii) Interquartile range: the range of the middle half of a distribution.
For data measured at an ordinal level, the range and interquartile range are the only appropriate measures of variability.
(iii) Standard deviation: average distance from the mean.
For more complex interval and ratio levels, the standard deviation and variance are also applicable.
(iv) Variance: average of squared distances from the mean.
For more complex interval and ratio levels, the standard deviation and variance are also applicable.
'''
```
