```python
## 1) Five plots that we can plot using the Seaborn library:
'''(i) Relational plots- scatterplot,lineplot . These plots are used for visualizing the statistical relationship between the data points. It allows the human to see trends and patterns in the data.
(ii) Distribution plots- histplot, kdeplot,ecdfplot,joinplot,distplot,pairplot,rugplot rug plot. It provides a high-level interface for drawing attractive and informative statistical graphics. 
It is is used for examining univariate and bivariate distributions.
(iii) Categorical plots- stripplot,swarmplot,boxplot,violinplot,pointplot,barplot. It is used when we deal with categorical data.'''
```

```python
## 2) To load the "fmri" dataset using the load_dataset function of seaborn:
import seaborn as sns
data=sns.load_dataset('fmri')
sns.lineplot(x=data.timepoint,y=data.signal)
```

```python
## 3) To load the "titanic" dataset :
import seaborn as sns
import matplotlib.pyplot as plt
data=sns.load_dataset("titanic")
sns.boxplot(x =data.pclass, y=data.age)
plt.show()
sns.boxplot(x =data.pclass, y=data.fare)
plt.show()
```

```python
## 4) To Use the "diamonds" dataset from seaborn:
import seaborn as sns
diamonds_data=sns.load_dataset('diamonds')
sns.histplot(diamonds_data.price)
sns.relplot(x=diamonds_data.cut,y=diamonds_data.price,hue=diamonds_data.color)
```

```python
## 5) To use the "iris" dataset from seaborn to plot a pair plot.
import seaborn as sns
data=sns.load_dataset('iris')
sns.pairplot(data)
sns.relplot(x=data.species,y=data.sepal_length,hue=data.petal_width)
```

```python
## 6) To use the "flights" dataset from seaborn to plot a heatmap:
import seaborn as sns
data=sns.load_dataset('flights')
data=data.pivot("month","year","passengers")
sns.heatmap(data)
```
