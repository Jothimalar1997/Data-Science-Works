##  1) Multiprocessing and its uses:

Multiprocessing is the ability of the system to handle multiple processess simultaneously and independently.
In multiprocessing system, the program is broken into smaller routines and the os gives processes to these routines for better performance.
With the help of multiproessing, the resources can be used efficiently and it enchances the performance of multicore machine.

## 2) Difference between multithreading and multiprocessing:

(i) Multithreading implements the concurrency while Multiprocessing implements the parallelism.
(ii) Python does not support multithreading incase of parallel computing. But multiprocessing can be achieved incase of parallel computing.
(iii) In multithreading, multiple threads at the same time are generated by a same process. In multiprocessing, multiple threads at the same time run across different cores.
(iv) Multithreading cannot be classified. Multiprocessing can be classified into symmetric and asymmetric.

## 3) To create a program using multiprocessing module:

```python
import multiprocessing
def square(num):
    print(f"square of the number is {num*num}")
def cube(num):
    print(f"cube of the number is {num*num*num}")
if __name__=="__main__":
    p1=multiprocessing.Process(target=square,args=(4,))
    p2=multiprocessing.Process(target=cube,args=(4,))
    p1.start()
    p2.start()
```

## 4) Multiprocessing pool:

(i) Multiprocessing pool allows us to create and manage process pools in python.
(ii) Process pool is a programming pattern for automatically managing a pool of worker processes.
(iii) The pool is responsible for fixed number of process-It controls when they are created based on their requirement. It also controls what they should do when they are not being used, such as making them wait without consuming computation resources.
(iv) multiprocessing.pool.Pool class provides a process pool in python.
(v) Create, Submit, Wait and Shitdown are the four main steps in the lifecycle of multiprocessing.pool class.

## 5) To create a pool of process using multiprocesssing module:

The pool process can be created using multiprocessing.pool.Pool class. The pool process can be configured by specifying arguments to multiprocessing.pool class constructor.
The most important argument is the 'processes' that specifies the number of worker child processes in the process pool.

```python
import multiprocessing
def square(num):
    print(f"Square of {num} is {num*num} ")
if __name__=="__main__":
    with multiprocessing.Pool(processes=5) as pool:
      out=pool.map(square,[1,2,3,4,5,6,7,8])
```

## 6) To create 4 processes that prints the different number using Multiprocessing module:

```python
import multiprocessing
def print_number(num):
    print(f"the value of {num} is {num*2} ")
if __name__=="__main__":
    with multiprocessing.Pool(processes=4) as pool:
      out=pool.map(print_number,[1,2,3,4,5,6,7,8])
```
